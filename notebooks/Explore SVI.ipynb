```python
import pandas as pd
import numpy as np
import os
import re
from fuzzywuzzy.process import extractOne,extractBests,extract
from fuzzywuzzy import fuzz
from pprint import pprint
```

```python
files = []
data = {}
for i in os.listdir('../data/raw/SVI/'):
    if ('SVI' in i)&('.csv' in i):
        year = re.findall('\d{4}',i)[0]
        df_year = pd.read_csv(f'../data/raw/SVI/{i}')
        df_year['year'] = int(year)
        
        data[f'{int(year)}']={'shape':df_year.shape,'columns':set(df_year.columns),'data':df_year}
        
        files.append(df_year)
        
```

```python
svi_dict_2018 = pd.read_excel('../data/raw/SVI/SVI2018_dictionary.xlsx')
svi_dict_2010  = pd.read_excel('../data/raw/SVI/SVI2010_dictionary.xlsx')

var_to_sec = {j['2018 VARIABLE NAME']:j['2018 DESCRIPTION'] for _,j in svi_dict_2018.iterrows()}
sec_to_var = {v:k for k,v in var_to_sec.items()}
```

```python
available_categories = var_to_sec

for idx,row in svi_dict_2010.iterrows():
    cat = extractBests(row['DESCRIPTION'],
                     available_categories.values(),
                     scorer = fuzz.token_sort_ratio,
                     limit=5)
    cat_2=[]
    for i in cat:
        cat_2.append(tuple(list(i)+[sec_to_var[i[0]]]))

    
    print(row['VARIABLE NAME'],row['DESCRIPTION'],'\n')
    pprint(cat_2)
    print('===================\n')
```
