```python
import pandas as pd
```

```python
df = pd.read_csv("../data/interim/SEPHER2.0.csv")
```

```python
df['FIPS'] = df['FIPS'].astype(int).astype(str).str.zfill(11)
```

```python
df_sample = df.sample(frac=1)
```

```python
state_mask = df_sample['FIPS'].apply(lambda x: int(x[0:2]))<=56
df_sample = df_sample[state_mask]
```

```python
all_nans = df_sample.isna().all()
all_nan_cols = all_nans.index[all_nans]
all_nan_cols
```

```python
df_sample.drop(all_nan_cols,axis=1,inplace=True)
```
